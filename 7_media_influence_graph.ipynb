```python
df2 = pd.read_csv("./data/edited_article_list(2024-06-03~).csv")
```

```python
df2.loc[df2['date'] == '19면 TOP', 'date'] = '2024.07.17.'
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))  # 그래프 크기 조정
sns.countplot(
    y = "Topic",  # y축에 Topic을 설정하여 가로 막대 그래프 생성
    palette = "pastel",
    data = df2
)
plt.tight_layout()  # 레이아웃 자동 조정
plt.show()
```
