# 제품별 키워드 딕셔너리 생성

```python
positive_keyword_categories = {
    'positive_taste':['delicious','good','great','diverse','addictive','amazing','tasty','refreshing','moist','nice','tender','chewy'],
    'portion':['large portion','portion','great','like','nicely','kind'],
    'hygiene':['clean'],
    'employee':['quick','quickly','kind'],
    'store':['plenty of space','spacious','affordable','reasonable','great value for money','clean']
}
```

```python
negative_keyword_categories = {
    'negative_taste':['disappointing','disappointed','terrible','tough','greasy','poor','worst','salty','burnt','not enough','bad','ordinary','unremarkable','not chewy','average','soggy'],
    'portion':['generous','small'],
    'hygiene':['dust','foreign objects','hygieneic','stains','condition'],
    'employee':['incorrect','arguing','upset','staff','response','yelled','insisted','friendliest','kind'],
    'store':['hot','warm','cold','air','waste of money','small','serious','cooler']
}
```

```python
file_names = ['translated_강서홈플러스점',
              'translated_고척동양대점',
              'translated_신촌2호점',
              'translated_우장산역점',
              'translated_성신여대점',
              'translated_교대역점',
              'translated_관악난곡점',
              'translated_목동남부시장점',
              'translated_홍대입구역점',
              'translated_구로디지털단지역점',
              'translated_수유역점',
              'translated_이수역점',
              'translated_신풍점',
              'translated_대흥역점',
              'translated_신림고시촌점'
]
```

```python
import pandas as pd
from tqdm import tqdm

# 키워드 카운팅 함수 (0인 값을 제외)
def count_keywords(text, keywords):
    counts = {keyword: text.lower().count(keyword) for keyword in keywords}
    return {k: v for k, v in counts.items() if v > 0}
    
# 모든 결과를 저장할 빈 리스트를 초기화
all_results = []

# 파일 리스트에 tqdm 적용
for file in tqdm(file_names, desc="Processing Files"):
    df = pd.read_csv(f"./data/bad_review_stores/{file}.csv", encoding='utf-8')
    results = {'title': file}  # 'title' 열을 가장 먼저 추가
    for category, keywords in positive_keyword_categories.items():
        df[category] = df['translated_review'].apply(lambda x: count_keywords(x, keywords))
        results[category] = df[category].apply(pd.Series).sum().to_dict()
    # 각 결과를 리스트에 추가
    all_results.append(results)
    
# 모든 결과를 데이터프레임으로 변환
final_df = pd.DataFrame(all_results)

# 결과를 하나의 CSV 파일로 저장
final_df.to_csv('./data/bad_review_stores/긍정적_키워드요약.csv', encoding="utf-8", index=False)
```

```python
import ast  # 문자열을 딕셔너리로 변환하기 위해 사용

# 각 딕셔너리 내 단어 수를 계산하는 함수
def count_words(d):
    # 딕셔너리인지 확인
    if isinstance(d, str):  # 만약 문자열이면, 이를 딕셔너리로 변환
        try:
            d = ast.literal_eval(d)  # 문자열을 딕셔너리로 변환
        except:
            return 0  # 변환이 실패하면 0 반환

    if isinstance(d, dict):  # 변환된 값이 딕셔너리라면 값들을 더함
        return sum(d.values())
    else:
        return 0  # 딕셔너리가 아니면 0 반환

# 함수 적용
df['taste_count'] = df['negative_taste'].apply(count_words)
df['portion_count'] = df['portion'].apply(count_words)
df['hygiene_count'] = df['hygiene'].apply(count_words)
df['employee_count'] = df['employee'].apply(count_words)
df['store_count'] = df['store'].apply(count_words)

# 필요한 컬럼만 남기기
df_counts = df[['title', 'taste_count', 'portion_count', 'hygiene_count', 'employee_count', 'store_count']]

# 결과 출력
print(df_counts)
```
